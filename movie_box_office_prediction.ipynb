```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from simple_linear_regression import SimpleLinearRegression
```

```python
df = pd.read_csv("datasets/clean.csv")

X = df.iloc[:, 0].values
y = df.iloc[:, 1].values

slr = SimpleLinearRegression()
slr.fit(X, y)

print('coef :', slr.coef_)
print('intercept :', slr.intercept_)
```

```python
plt.figure(figsize= (16, 9))
plt.scatter(X,y, alpha= 0.5)
plt.plot(X, slr.predict(X), color= 'red', linewidth= 5)
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.xlabel("Production Budget (in USD)")
plt.ylabel("Worldwide Gross (in USD)")
plt.title("Film Cost vs Global Revenue")
plt.show()
```
